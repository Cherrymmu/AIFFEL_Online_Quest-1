```python
import random as r

class Account:
    bank_name = "SC은행"
    account_count = 0

    def __init__(self):
        self.name = []
        self.balance = []
        self.account_num = []
        self.deposit_count = 0

    def create(self):
        while True:
            c = input("어서오세요. 신규 계좌 등록을 진행하겠습니다.\n진행하시려면 Y를 종료하시려면 N을 입력해주세요: \n >>> ")
            if c == "Y".lower():
                account_name = input("예금주명을 입력해주세요: ")
                account_balance = int(input("초기잔액을 입력해주세요: "))
                # 계좌번호 랜덤하게 받기
                account_num = self.generate_account_num()

                self.name.append(account_name)
                self.balance.append(account_balance)
                self.account_num.append(account_num)

                # 계좌가 생성될 때마다 계좌 개수 증가
                Account.account_count += 1

            elif c == "N".lower():
                print("계좌 개설을 종료합니다.")
                break
    def generate_account_num(self):
        first_part = str(r.randint(100, 999))
        second_part = str(r.randint(11, 99))
        third_part = str(r.randint(100000, 999999))
        account_number = f'{first_part}-{second_part}-{third_part}'
        return account_number

    # 생성된 계좌의 개수를 출력
    def get_account_num(self):
        print(f"생성된 계좌의 개수는 {self.account_count}개 입니다.")
        print(f'현재정보는 은행이름: {Account.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_num}, 잔고: {self.balance}')
    # 입금을 위한 메서드
    def deposit(self):
        while True:
            d = int(input("입금을 하시겠습니까?\n입금하시려면 1 을 다음으로 넘어가시려면 2를 >>> "))
            if d == 1:
                select_account_num = input("입금하실 계좌번호를 입력해주세요: ")
                if select_account_num in select_account_num:
                    index = self.account_num.index(select_account_num)
                    amount = int(input("입금할 금액을 입력해주세요: "))
                    if amount >= 1:
                        self.balance[index] += amount
                        self.deposit_count += 1
                        print(f'{self.name[index]}님, {amount}원이 성공적으로 입금되었습니다.')

                        # 이자 지급 조건 확인 (입금 횟수가 5회일 때)
                        if self.deposit_count == 5:
                            interest_rate = 0.01  # 1% 이자율
                            interest_amount = int(self.balance[index] * interest_rate)
                            self.balance[index] += interest_amount
                            print(f'이자 {interest_rate * 100}%가 지급되었습니다. 현재 잔고: {self.balance[index]}원')
                    else:
                        print("입금 금액은 최소 1원 이상이어야 합니다. 다시 시도해주세요.")
                else:
                    print("해당하는 계좌번호가 없습니다. 다시 입력해주세요.")
            elif d == 2:
                print("입금을 종료합니다.")
                break

    # 출금을 위한 메서드
    def withdraw(self):
        while True:
            w = int(input("출금을 하시겠습니까?\n출금하시려면 1을, 다음으로 넘어가시려면 2를 >>> "))
            if w == 1:
                select_account_num = input("출금하실 계좌번호를 입력해주세요: ")
                if select_account_num in self.account_num:
                    index = self.account_num.index(select_account_num)
                    amount = int(input("출금할 금액을 입력해주세요: "))
                    if amount > 0 and amount <= self.balance[index]:
                        self.balance[index] -= amount
                        print(f'{self.name[index]}님, {amount}원이 성공적으로 출금되었습니다. 현재 잔고: {self.balance[index]}원')
                    elif amount <= 0:
                        print("출금 금액은 0보다 큰 금액이어야 합니다. 다시 시도해주세요.")
                    else:
                        print("잔고가 부족하여 출금할 수 없습니다. 다시 시도해주세요.")
                else:
                    print("해당하는 계좌번호가 없습니다. 다시 입력해주세요.")
            elif w == 2:
                print("출금을 종료합니다.")
                break

    # 정보 출력 메서드
    def display_info(self):
        formatted_balance = "{:,}".format(self.balance)
        print(f'현재정보는 은행이름: {Account.bank_name}, 예금주: {self.name}, 계좌번호: {self.account_num}, 잔고: {formatted_balance}' )




acc = Account()
acc.create()
acc.get_account_num()
acc.deposit()
acc.withdraw()
acc.display_info()
```
